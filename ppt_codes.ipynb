```python
n = int(input("Enter range: "))
a = list(map(int, input("Enter elements: ").split()))
k = 0
for i in range(0,n-1):
    if a[i] > a[i + 1]:
        k += 1

if k == 0:
    print("sorted")
else:
    print("not sorted")
```

```python
n = int(input("Enter range: "))
a = list(map(int, input("Enter elements: ").split()))
b = list(map(int, input("Enter elements: ").split()))
k=0
for i in range(n):
  if a[i]==b[i]:
    k+=1
if k==1:
  print("same")
else:
  print("diff")
```
